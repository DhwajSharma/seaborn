```python
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
```

```python
df.head()
```

```python
df.dtypes
```

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

    Jointplot

Univariate Analysis

```python
sns.jointplot(x = 'tip', y = 'total_bill', data = df, kind = 'hex')
```

```python
sns.jointplot(x = 'tip', y = 'total_bill', data = df, kind = 'reg')
```

Pair plot

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue = 'sex')
```

Dist plot

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde = False, bins = 10)
```

Categorical plots : boxplot, violinplot, countplot, barplot

```python
# Count Plot
sns.countplot('sex', data = df)
# sns.countplot(y = 'sex', data = df)
```

```python
# BoxPlot
sns.boxplot('smoker', 'total_bill', data = df, palette = 'rainbow')
```

```python
sns.boxplot(x = 'total_bill', y = 'day', hue = 'smoker', data = df)
```

```python
# Violin plot
sns.violinplot(x = 'total_bill', y = 'day', data = df, palette = 'rainbow')
```
